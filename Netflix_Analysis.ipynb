### Question 10

<b>Q10.1 What is the maximum duration of a Movie on Netflix? </b>

```python
df.head(1)
```

```python
#Since duration column is str type and uneven spacing 
#Therefore, we use str.extract to extract numeric part
df['duration_numeric'] = df['duration'].str.extract(r'(\d+)')

# Fill NaN values with 0 and convert to integer
df['duration_numeric'] = df['duration_numeric'].fillna(0).astype(float)
```

```python
df[df['type'] == 'Movie']['duration_numeric'].max()
```

<b>Q10.2 What is the maximun duration of TV Show in Netflix?</b>

```python
df[df['type'] == 'TV Show']['duration_numeric'].max()
```

<b>Alternate Solution for Both the Problems</b>

```python
result = df.groupby('type')['duration_numeric'].max().reset_index()
result
```
